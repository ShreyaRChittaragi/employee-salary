```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler,OneHotEncoder

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

models = {
    'Logistic Regression': LogisticRegression(),
    'Random Forest': RandomForestClassifier(),
    'GradientBoosting': GradientBoostingClassifier(),
    'KNN': KNeighborsClassifier(),
    'SVM': SVC()
}

results = {}

for name,model in models.items():
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('model', model)
    ])
    pipe.fit(x_train, y_train)
    y_pred = pipe.predict(x_test)
    acc=accuracy_score(y_test,y_pred)
    results[name] = acc
    print(f"{name} Accuracy: {acc}")
    print(classification_report(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier

# Add new models to the dictionary
new_models = {
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Gaussian NB': GaussianNB(),
    'LightGBM': LGBMClassifier(random_state=42),
    'XGBoost': XGBClassifier(random_state=42),
    'AdaBoost': AdaBoostClassifier(random_state=42),
    'GradientBoosting': GradientBoostingClassifier(random_state=42) # Added GradientBoosting here as well for completeness
}

# Merge the new models with the existing ones for comprehensive evaluation
# Ensure no duplicates if GradientBoosting was already in 'models'
models.update(new_models)

print("Updated list of models to evaluate:")
print(list(models.keys()))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report

results = {}

for name, model in models.items():
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('model', model)
    ])
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results[name] = acc
    print(f"{name} Accuracy: {acc}")
    print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

# Define hyperparameter grids for the top models
param_grids = {
    'LightGBM': {
        'n_estimators': [100, 200, 300],
        'learning_rate': [0.01, 0.05, 0.1],
        'num_leaves': [31, 50, 100]
    },
    'XGBoost': {
        'n_estimators': [100, 200, 300],
        'learning_rate': [0.01, 0.05, 0.1],
        'max_depth': [3, 5, 7]
    },
    'GradientBoosting': {
        'n_estimators': [100, 200, 300],
        'learning_rate': [0.01, 0.05, 0.1],
        'max_depth': [3, 5, 7]
    }
}

# Select the top performing models for tuning
# Based on previous evaluation, LightGBM, XGBoost, and GradientBoosting performed well
top_models = {
    'LightGBM': models['LightGBM'],
    'XGBoost': models['XGBoost'],
    'GradientBoosting': models['GradientBoosting']
}

tuned_results = {}

for name, model in top_models.items():
    print(f"Tuning {name}...")
    grid_search = GridSearchCV(estimator=model, param_grid=param_grids[name], cv=5, scoring='accuracy')
    grid_search.fit(X_train, y_train)

    print(f"Best hyperparameters for {name}: {grid_search.best_params_}")

    best_model = grid_search.best_estimator_
    y_pred_tuned = best_model.predict(X_test)

    acc_tuned = accuracy_score(y_test, y_pred_tuned)
    tuned_results[name] = acc_tuned

    print(f"Tuned {name} Accuracy: {acc_tuned}")
    print(classification_report(y_test, y_pred_tuned))
    print("-" * 50)
```

```python
# Combine results from initial evaluation and tuning
all_results = results.copy()
all_results.update(tuned_results)

# Find the model with the highest accuracy
best_model_name = max(all_results, key=all_results.get)
best_accuracy = all_results[best_model_name]

print(f"The best performing model is: {best_model_name} with an accuracy of {best_accuracy:.4f}")
```

```python
import joblib
from sklearn.ensemble import GradientBoostingClassifier

# Retrain and save the best model
best_model = GradientBoostingClassifier(**grid_search.best_params_)
best_model.fit(X_train, y_train)
joblib.dump(best_model, "salary_predictor_model.pkl")

# Assuming 'encoders' are already saved from previous steps
# joblib.dump(encoders, "label_encoders.pkl")

with open("app.py", "w") as f:
    f.write('''
import streamlit as st
import pandas as pd
import joblib

# Load the best model and encoders
model = joblib.load("salary_predictor_model.pkl")
encoders = joblib.load("label_encoders.pkl")

st.set_page_config(page_title="Employee Salary Predictor", page_icon="🧑‍💼")

st.title("🧑‍💼 Employee Salary Predictor")
st.markdown("Predict whether an employee's salary is >50K or <=50K based on their attributes.")

st.header("Employee Information")

# Organize inputs into columns for better layout
col1, col2 = st.columns(2)

with col1:
    age = st.slider("Age", 17, 90, 30)
    workclass = st.selectbox("Workclass", encoders['workclass'].classes_)
    # education = st.selectbox("Education", encoders['education'].classes_) # Removed as dropped
    marital_status = st.selectbox("Marital Status", encoders['marital-status'].classes_)
    occupation = st.selectbox("Occupation", encoders['occupation'].classes_)
    race = st.selectbox("Race", encoders['race'].classes_)
    capital_gain = st.number_input("Capital Gain", 0, 100000, 0, help="Amount of capital gains")


with col2:
    fnlwgt = st.number_input("fnlwgt", 10000, 1000000, 150000, help="Final weight (sampling weight)")
    educational_num = st.slider("Education Number", 1, 16, 10, help="Numerical representation of education level")
    relationship = st.selectbox("Relationship", encoders['relationship'].classes_)
    gender = st.selectbox("Gender", encoders['gender'].classes_)
    hours_per_week = st.slider("Hours per Week", 1, 100, 40)
    native_country = st.selectbox("Country", encoders['native-country'].classes_)
    capital_loss = st.number_input("Capital Loss", 0, 5000, 0, help="Amount of capital losses")


input_data = {
    "age": [age],
    "workclass": [workclass],
    "fnlwgt": [fnlwgt],
    # "education": [education], # Removed as dropped
    "educational-num": [educational_num],
    "marital-status": [marital_status],
    "occupation": [occupation],
    "relationship": [relationship],
    "race": [race],
    "gender": [gender],
    "capital-gain": [capital_gain],
    "capital-loss": [capital_loss],
    "hours-per-week": [hours_per_week], # Corrected feature name
    "native-country": [native_country]
}

input_df = pd.DataFrame(input_data)

# Encode categorical variables using the loaded encoders
for column in input_df.select_dtypes(include='object').columns:
     if column in encoders:
        try:
            input_df[column] = encoders[column].transform(input_df[column])
        except ValueError as e:
            st.error(f"Error encoding column '{column}': {e}. Please select a valid category.")
            st.stop() # Stop execution if encoding fails


if st.button("Predict Salary Category"):
    result = model.predict(input_df)[0]
    prediction = encoders['income'].inverse_transform([result])[0]

    if prediction == ">50K":
        st.success(f"💰 Predicted Salary Category: {prediction} 🎉")
    else:
        st.info(f"💰 Predicted Salary Category: {prediction}")

st.markdown("---")
st.markdown("Built with ❤️ using Streamlit and Scikit-learn")

''')
```
